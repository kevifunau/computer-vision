```python
import cv2
import numpy as np
import sys
import os
import matplotlib.pyplot as plt
```

### cv2.imread()
Second argument is a flag which specifies the way image should be read.

- cv2.IMREAD_COLOR : Loads a color image. Any transparency of image will be neglected. It is the default flag.
- cv2.IMREAD_GRAYSCALE : Loads image in grayscale mode
- cv2.IMREAD_UNCHANGED : Loads image as such including alpha channel

```python
# load image at greyscale mode
image = cv2.imread("ductile_iron2-0.jpg",cv2.IMREAD_GRAYSCALE)
print(image,image.shape)
```

```python
# define 0-255 array
hist_array = np.zeros(256,dtype=np.int32)
for i in image.ravel():
    hist_array[i]+=1
# transfer to frequency [0,1]
hist_array_f = np.zeros(256,dtype=np.float128)
for i in range(256):
    hist_array_f[i] = hist_array[i] / image.size
```

```python
hist_array_f
```

```python
max_th=0
max_th_value = 0
rr = []
for t in range(256):
    # probability of foreground
    wf = sum([ p for p in hist_array_f[:t]])
    wb = sum([ p for p in hist_array_f[t:]])
    # varience 
    vf = 0
    vb = 0
    for i in range(t):
        vf += i*hist_array_f[i]
    for i in range(t,256):
        vb += i*hist_array_f[i]
        
    if(wf ==0 or wb ==0):
        continue
        
    vf = vf / wf 
    vb = vb / wb 
    
    # the inter-class varience
    r = wf*wb*(vf - vb)*(vf - vb)
    
    
    rr.append([r,t])
    
    if r > max_th_value:
        max_th_value = r
        max_th = t

print(max_th)
    
    
    
    
```
